It is easy to get the marginal condition as: $0<x<1$, also $y=z-x>0$ and thus $z>x$
<br> Because 
\begin{align}
\frac{\sigma(x, y)}{\sigma(x, z)} = 
\begin{bmatrix}
    \frac{\sigma(x)}{\sigma(x)} & \frac{\sigma(x)}{\sigma(z)}  \\
    \frac{\sigma(y)}{\sigma(x)}  & \frac{\sigma(y)}{\sigma(z)}     
\end{bmatrix} 
= \frac{\sigma(y)}{\sigma(z)}=1
\end{align}

<br>Then, we need to prove $\iint f(x, z-x)dxdz = 1$
<br>
\begin{equation}
\displaystyle\int_{0}^{1} dx \displaystyle \int_{x}^{\infty} f(x, z-x)dz = \displaystyle\int_{0}^{1} dx \displaystyle \int_{x}^{\infty} e^{x-z} dz = 1
\end{equation}
<br>Therefore, it integrates to 1


Assume the median is b, then we have
\begin{equation}
    \int_{ 0 }^{ + 1 }dx \int_{x}^{b}{f(x, z-x)dz} = \int_{ 0 }^{ + 1 }dx \int_{x}^{b}{e^{x-z}} = 0.5
\end{equation}
Then we could easily solve $ b = \frac{0.5}{e-1} = 1.23 $

If we want to sketch the density of Z, then we need to calculate:
\begin{equation}
   f(z) =  \int_{  0 }^{ 1 } e^{x-z} dx= e^{1-z}-e^{-z}, z>1
\end{equation}
\begin{equation}
   f(z) =  \int_{  0 }^{ z } e^{x-z} dx= 1-e^{-z}, 0<z<1
\end{equation}

```python
# sketch the density
import matplotlib.pyplot as plt
import numpy as np

# 1000 linearly spaced numbers
Z = np.linspace(0,100,1000)
y = []
# the function f(z), which is calculated before
for z in Z:
    if z>1:
        y.append(np.exp(1-z)-np.exp(-z))
    else:
        y.append(1-np.exp(-z))

# setting the axes at the centre
plt.xlabel('z')
plt.ylabel('f(z)')

# plot the function
plt.plot(Z,y)

# show the plot
plt.show()
```

### Problem 4

Given a multivariate Gaussian distribution $\mathbf{x} \sim N(\mathbf{\mu}, \Sigma)$ where:
\begin{align}
\mathbf{\mu} = 
\begin{bmatrix}
    \mu_{1}  \\
    \mu_{2}      
\end{bmatrix}
\end{align}
and 
\begin{align}
\mathbf{\Sigma} = 
\begin{bmatrix}
    \Sigma_{11} & \Sigma_{12}  \\
    \Sigma_{21}  & \Sigma_{22}     
\end{bmatrix} 
\end{align}
The conditional distribution of the variable $x_{1}$ given $x_{2}$ of the above should be:
\begin{align}
\mu_{1|2} = \mu_{1} + \Sigma_{12}\Sigma_{22}^{-1}(x_{2} - \mu_{2})\\
\Sigma_{1|2} = \Sigma_{11} - \Sigma_{12}\Sigma_{22}^{-1}\Sigma_{21}\\
\end{align}
Thus: $x_{1}\sim N(\mu_{1|2}, \Sigma_{1|2})$

In this case, we can easily compute when $y = -0.5$, $x\sim N(2.15, 0.55)$.
It is also easy to get the marginal distribution of $x$ is $x\sim N(2, 1)$

```python
# Preparation
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy import stats

mean = [2, 0]
cov = [[1., -1.5], [-1.5, 5]]
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, 'o')
plt.show()
```

```python
# for p(x, y)

def p1(x,y):
    mean = [2, 0]
    cov = [[1, -1.5], [-1.5, 5]]
    X = np.array([x, y])
    return multivariate_normal.pdf(X, mean=mean, cov=cov)

xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()

X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
prob1 = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        prob1[i, j] = p1(X[i,j], Y[i,j])

positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([x, y])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, X.shape)

fig, ax = plt.subplots()
ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r, extent=[xmin, xmax, ymin, ymax])
ax.plot(x, y, 'k.', markersize=2)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
plt.show()

fig, ax = plt.subplots()
ax.imshow(np.rot90(prob1), cmap=plt.cm.gist_earth_r, extent=[xmin, xmax, ymin, ymax])
ax.plot(x, y, 'k.', markersize=1)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
plt.show()
```

```python
# for p(x)
x = np.random.normal(2, 1, 1000)
xmin, xmax = x.min(), x.max()
X = np.linspace(xmin, xmax, 100)[:,None]
positions = X.flatten()
values = x.flatten()
gkde = stats.gaussian_kde(values)
KDE = gkde.evaluate(positions)

mean = 2
std = 1
x.sort()
y= 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))


ax = plt.figure()
plt.plot(x, y, 'b-', label = 'Exact')
plt.plot(X, KDE, 'r-', label = 'Prediction')
plt.legend(loc='upper left', frameon=False, prop={'size': 13})
plt.show()
```

```python
# preparation
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mean = [3, -1]
cov = [[1.2, 0.4], [0.4, 1.0]]
```

```python
# generate 5000 realizations of X
x1, x2 = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x1, x2, 'X')
plt.show()
```

```python
# generate 5000 samples of Y
alpha = 1.5
beta = 0.2
y1 = alpha * x1
y2 = x2 / alpha + beta * (x1**2 + alpha**2)
plt.plot(y1, y2, 'Y')
plt.show()
```

```python
# plot the samples of Y along with a Gaussian Kernel density estimator
xmin = y1.min()
xmax = y1.max()
ymin = y2.min()
ymax = y2.max()

# Perform a kernel density estimate on the data
X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([y1, y2])
kernel = stats.gaussian_kde(values)
Z = np.reshape(kernel(positions).T, X.shape)

# Plot the results:
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.imshow(np.rot90(Z), cmap=plt.cm.gist_earth_r, extent=[xmin, xmax, ymin, ymax])
ax.plot(y1, y2, 'k.', markersize=0.2)
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
plt.show()
```
